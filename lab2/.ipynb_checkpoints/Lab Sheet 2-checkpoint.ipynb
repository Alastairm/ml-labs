Here we will read the training.csv & testing.csv, and remove all 'pred_minus_obs' data columns as per instruction 2.

## 3. Checking label balance...

```python
def count_labels(data: List[str], class_labels: List[str]) -> List[int]:
    """Counts number of instances of each class label in data."""
    counts = []
    for label in class_labels:
        counts.append(sum([x == label for x in data]))
    return counts

def percents_from_counts(class_label_counts: List[int]) -> List[float]:
    """Converts label counts to percentages."""
    total = sum(class_label_counts)
    percents = []
    for x in class_label_counts:
        percents.append(f"{(x/total*100):4.1f}%")
    return percents

training_counts = count_labels(training["class"], CLASS_LABELS)
testing_counts = count_labels(testing["class"], CLASS_LABELS)
print("Class totals:")
print(pd.DataFrame([training_counts, testing_counts],
             ["training", "testing"],
             CLASS_LABELS))

train_percents = percents_from_counts(training_counts)
test_percents = percents_from_counts(testing_counts)
print("\nClass proportions:")
print(pd.DataFrame([train_percents, test_percents],
             ["training", "testing"],
             CLASS_LABELS))
```

The training set has more than double the proportion of class "h" as the test set, but Du says this is fine.

## 4. Data scaling and normalisation


```python
# Scaling of sample features
from sklearn.preprocessing import StandardScaler
from sklearn.exceptions import DataConversionWarning
import warnings
# We don't need to be told ints are turning into floats...
warnings.filterwarnings("ignore", category=DataConversionWarning)

# Scale sample feature values
training_samples = training.drop("class", axis=1)
testing_samples = testing.drop("class", axis=1)

scaler = StandardScaler().fit(training_samples)

training_samples = scaler.transform(training_samples)
testing_samples = scaler.transform(testing_samples)


# Calculate binary class membership
training_labels = {}
testing_labels = {}

for target_class in CLASS_LABELS:
    training_labels[target_class] = [x is target_class for x in training["class"]]
    testing_labels[target_class] = [x is target_class for x in testing["class"]]
```

## 5. Stochastic Gradient Descent

[Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDRegressor.html#sklearn.linear_model.SGDRegressor)

```python
import numpy as np
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import confusion_matrix

def print_confusion_matrix(confusion_matrix: ndarray, label: str) -> None:
    """Print a confusion matrix with labeled rows and columns."""
    tn, fp, fn, tp = confusion_matrix.ravel()
    print(f"Class {label} confusion matrix:")
    print(pd.DataFrame([[tp, tn], [fp, fn]],
                 ["Corrrect", "Incorrect"],
                 [f"is {label}", f"not {label}"]))

def binary_classify(classifier: SGDClassifier) -> Dict:
    pass

# Create classifier
clf = SGDClassifier(loss="log",
                    penalty="l2",
                    max_iter=50,
                    tol=0.01)

sgd_results = {}
for label in CLASS_LABELS:
    # Classification fit & predict parameters
    x_train = training_samples
    y_train = training_labels[label]
    x_test = testing_samples
    # Get true and predicted labels
    y_true = testing_labels[label]
    y_pred = clf.fit(x_train, y_train).predict(x_test)
    # Store results
    sgd_results[label] = {}
    sgd_results[label]["y_true"] = y_true
    sgd_results[label]["y_pred"] = y_pred
    sgd_results[label]["cm"] = confusion_matrix(y_true, y_pred)

total_cm = [[0,0],[0,0]]
print("Stochastic gradient descent (SGD) binary classification:")
for label in CLASS_LABELS:
    print("")
    print_confusion_matrix(sgd_results[label]["cm"], label)
    total_cm = np.add(total_cm, sgd_results[label]["cm"])
    
#print_confusion_matrix(total_cm, "all")
#tn, fp, fn, tp = total_cm.ravel()
#print(f"\naccuracy:{(tn+tp)/1300}")
```
